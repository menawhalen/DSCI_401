```python
batting = pd.read_csv("https://raw.githubusercontent.com/menawhalen/DSCI_401/refs/heads/main/data/Batting.csv")
people = pd.read_csv("https://raw.githubusercontent.com/menawhalen/DSCI_401/refs/heads/main/data/People.csv")
```

```python
batting_full = pd.merge(batting, people, left_on = "playerID", right_on = "playerID", how = "left")
```

```python
batting_full.head()
```

```python
batting_full["season_age"] = batting_full["yearID"] - batting_full["birthYear"]
```

```python
batting_full["season_age"]
```

```python
manny = batting_full.loc[batting.playerID == "ramirma02"]
```

```python
manny.loc[manny["HR"] > 30][["yearID", "HR", "season_age"]].sort_values(by = "yearID", ascending = False)
```

```python
batting_full.groupby("birthCountry")["H"].agg(["sum"]).sort_values("sum", ascending = False)
```

```python
country_count = batting_full.groupby("birthCountry")["playerID"].nunique().reset_index(name="n")
```

```python
one_country = country_count[country_count.n == 1]
one_country.head()
```

```python
country_people = batting_full.groupby(["birthCountry", "playerID"])["yearID"].nunique().reset_index(name="n")
country_people.head()
```

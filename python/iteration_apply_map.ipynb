## Vectorization with np.vectorize
In Python, `vectorization` is a technique that allows you to apply a function to a whole array or list of values at once, rather than using loops. This approach is not only more efficient but also takes advantage of optimized libraries like numpy for faster computation. Vectorization is especially helpful when you want to perform operations on large datasets without manually iterating over each element.

### Defining and Vectorizing a Function
1. Define the function: We start with a basic function f that returns the square root of a single value using np.sqrt.

2. Vectorize the function: We then use np.vectorize(f) to create a vectorized version of the function, stored in vf. The vectorized function can now take in lists or arrays and return the square root of each element in a single operation.

3. Apply the vectorized function: Now, we can pass an entire list (e.g., [1, 2, 3]) to vf, which returns an array with the square root of each element.

````python
vf = np.vectorize(f)
vf([1,2,3])
````

## Using apply in Python with Pandas

### Example: Applying a Function Across Columns or Rows
Suppose we have a DataFrame Teams with several numeric columns, and we want to calculate the mean for each column within a specific range of columns.

Using .apply(): We use .apply(np.mean, axis=0) to calculate the mean of each column.

- np.mean is the function applied to each selected column.
- axis=0 specifies that the function should be applied across each column.
- Setting axis=1 would apply the function across rows instead, calculating the mean for each row.

## Using Lambda Functions with apply
Lambda functions are a way to create quick, one-line functions for a specific task without defining a full function elsewhere. Here, theyâ€™re used in combination with `.apply()` to do something specific for each group of data, like selecting the "first" or "last" entry. The point of learning this is to practice creating small, task-specific functions to operate on data. Lambda functions help make code shorter and more readable when the task is straightforward.

### Applying Functions to Groups of Data
This is where `.apply()` shines because it allows you to take action on each group separately. The idea here is to see that you can split data up and apply different logic to each section as needed. Itâ€™s a foundation for many data tasks because it mimics real-world organization: you often need to analyze or report on data by category, date, or group.

````python
angels = Teams[Teams.franchID == "ANA"].groupby(["teamID","name"])


first = angels.apply(lambda x: x.head(1))[["name","yearID"]]
first.reset_index(drop=True, inplace=True) ## needed for formatting

last = angels.apply(lambda x: x.tail(1))[["name","yearID"]]
last.reset_index(drop=True, inplace=True)

merg = first.merge(last, on = "name")
merg = merg.rename(columns = {"yearID_x":"start","yearID_y":"end"})
print(merg)
````

## Map
The map function in Python is a built-in function that allows you to apply a specified function to each item in an iterable (like a list or a set) and returns a new iterator with the results. Itâ€™s especially useful when you want to apply the same operation to every element in a collection without writing a loop.
```
map(function, iterable)
```
- function: This is the function you want to apply to each element in the iterable. It can be a regular function or a lambda function (an anonymous function).
- iterable: This is the sequence of items (like a list, tuple, or set) that map will apply the function to.

### How It Works:
1. Function Application: map takes each element from the iterable and applies the specified function to it.
2. Returns an Iterator: Instead of directly returning a new list, map returns a special object called an iterator. This is memory efficient, especially for large datasets, because it computes each result only when needed.
3. Casting to a List: Since map returns an iterator, you often see it wrapped in list() when you want all results at once, as in list(map(...)).

Example with Basic Functions
Hereâ€™s a simple example to illustrate how map works with a function that squares each element in a list:

````python
# Define a simple function to square a number
def square(x):
    return x ** 2

# Apply the function to each item in the list
numbers = [1, 2, 3, 4]
squared_numbers = map(square, numbers)

# Convert map object to list to see the results
print(list(squared_numbers))
squared_numbers
````

### Using map with a Lambda Function
For small, quick operations, you can use a lambda function inside map instead of defining a separate function. Hereâ€™s the previous example rewritten using a lambda:



````python
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))
````

## map v apply in python
- map: A built-in Python function primarily designed to apply a single function to each element in a Python iterable (e.g., list, tuple). map returns a map object (an iterator) that generates results lazily, making it memory efficient.

- apply: A method available in the Pandas library, designed specifically for applying functions to either rows or columns of a DataFrame or Series. It is widely used for DataFrame manipulations where the function requires context on each row or column.

### Typical Usage

- `map`: Typically used for transforming lists, dictionaries, or similar collections outside of Pandas. Since it works with individual elements, itâ€™s best suited for element-wise transformations and works well with simpler operations that donâ€™t depend on DataFrame context.

- `apply`: Often used within a DataFrame to apply operations that might require row or column context. You can specify axis=0 to apply functions across columns or axis=1 to apply across rows, making it highly flexible for row-wise or column-wise transformations.

````python
# Example with a list
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # [1, 4, 9, 16]
````

````python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [10, 20, 30]})
df['A_squared'] = df['A'].apply(lambda x: x ** 2)
print(df)
````

### Performance and Efficiency
- `map`: Faster for applying functions to basic Python data structures due to its native iterator-based design. Ideal for simple element-wise operations that donâ€™t require complex logic or row/column context.

- `apply`: Can be slower with large DataFrames because itâ€™s more flexible and provides row or column context. However, it is highly compatible with DataFrames and can handle complex operations.

### Return Type
- `map`: Returns an iterator (which can be converted to a list or other collection), meaning it is memory efficient for large datasets when used without conversion to a list.

- `apply`: Returns a Pandas Series or DataFrame, depending on whatâ€™s applied and the axis specified. This means it integrates seamlessly with other DataFrame methods and operations.

### Common use cases

- `map`: Transforming elements of a list, dictionary keys/values, or Series with simple, element-wise functions. In Pandas, map is also used to map values in a Series based on a dictionary or function.

- `apply`: Applying row or column-based transformations that require context, or aggregating functions that may need additional arguments, such as lambda functions that compute row-specific operations.


````python
# Example: Using map with a dictionary in Pandas
df['A_mapped'] = df['A'].map({1: 'one', 2: 'two', 3: 'three'})
print(df.A_mapped)
````

````python
# Example: Summing values in each row
df['row_sum'] = df.apply(lambda row: row['A'] + row['B'], axis=1)
print(df.row_sum)
````
